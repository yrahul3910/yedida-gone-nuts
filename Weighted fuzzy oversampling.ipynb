```python
from raise_utils.transforms import Transform
from raise_utils.data import Data
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from raise_utils.learners import FeedforwardDL
from raise_utils.hyperparams import BinaryGHOST
from mlxtend.plotting import plot_decision_regions
from matplotlib import pyplot as plt
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, weights=(.75,))
```

```python
data = Data(*train_test_split(X, y))
```

```python
plt.scatter(X.T[0], X.T[1], c=y);
```

```python
learner = FeedforwardDL(n_units=2, n_layers=2, n_epochs=30)
learner.set_data(*data)
```

```python
learner.fit()
```

```python
plot_decision_regions(data.x_train, data.y_train, learner)
```

```python
learner = FeedforwardDL(weighted=True, n_units=2, n_layers=2, n_epochs=30)
learner.set_data(*data)
```

```python
learner.fit()
```

```python
plot_decision_regions(data.x_train, data.y_train, learner)
```

```python
learner = BinaryGHOST(['f1'], n_runs=1, ultrasample=False, autoencode=False)
learner.set_data(*data)
```
