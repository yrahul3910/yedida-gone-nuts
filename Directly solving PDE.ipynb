```python
from raise_utils.learners import Autoencoder
from raise_utils.data import DataLoader
import numpy as np
```

```python
data = DataLoader.from_files(base_path='../Dodge/data/defect/', files=['synapse-1.0.csv', 'synapse-1.1.csv'])
```

```python
autoencoder = Autoencoder(n_layers=2, n_units=[10, 5], n_out=2)
autoencoder.set_data(*data)
autoencoder.fit()
```

```python
encoded_train = autoencoder.encode(data.x_train.to_numpy())
encoded_test = autoencoder.encode(data.x_test.to_numpy())
```

```python
from sympy.abc import p, q
from sympy import Lambda, stationary_points, pprint
from sympy.solvers.ode import dsolve
import sympy as sp
```

```python
def hypothesis(x, x_mu):
    return 1. / (1. + sp.exp(-(p * (x[0]-x_mu[0]) + q * (x[1]-x_mu[1]))))
```

```python
def loss_fn(y, y_hat):
    return -y * sp.log(y_hat) - (1. - y) * sp.log(1. - y_hat)
```

```python
def cost_fn(x_train, y_train):
    x_mu = np.mean(x_train, axis=0)
    assert x_mu.shape[0] == 2
    
    loss = None
    for i in range(len(x_train)):
        yhat = hypothesis(x_train[i], x_mu)
        if loss is None:
            loss = yhat
        else:
            loss = loss + loss_fn(y_train[i], yhat)
    
    return loss
```

```python
loss = cost_fn(encoded_train, data.y_train)
```
